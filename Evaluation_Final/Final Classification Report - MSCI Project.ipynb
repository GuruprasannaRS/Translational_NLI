```python
df_org = pd.read_csv(r'D:\UWaterloo\MSCI_641_Text_Analytics\Project\Output\Original\metrics_seven_fo_br.csv',
                header=None, names=['Evaluation', 'Scores'])
df_org[['bert_org', 'roberta_org']] = df_org['Scores'].str.split(', ', expand=True)
df_org['bert_org'] = df_org['bert_org'].str.replace('[', '')
df_org['roberta_org'] = df_org['roberta_org'].str.replace(']', '')
df_org['bert_org'] = pd.to_numeric(df_org['bert_org'])
df_org['roberta_org'] = pd.to_numeric(df_org['roberta_org'])
df_org.drop(columns=['Scores'], inplace=True)
new_values = ['train_accuracy', 'val_accuracy', 'test_accuracy','precision','recall','f1','kappa']

df_org['Evaluation'] = new_values
df_org['bert_org'] = df_org['bert_org'].round(3)
df_org['roberta_org'] = df_org['roberta_org'].round(3)
df_org
```

```python
final_df = df_mnb_pred.merge(df_mnb_org, on='Evaluation').merge(df_cnn_pred, on='Evaluation').merge(df_cnn_org, on='Evaluation').merge(df_p, on='Evaluation').merge(df_o, on='Evaluation').merge(df_pred, on='Evaluation').merge(df_org, on='Evaluation') 
```

```python
final_df_org = final_df.drop(['mnb_pred','cnn_pred','electra_pred','xlnet_pred','bert_pred','roberta_pred'], axis = 1)
final_df_org.columns = ['Evaluation','mnb', 'cnn', 'electra', 'xlnet', 'bert', 'roberta']
final_df_org
```

```python
final_df_pred = final_df.drop(['mnb_org','cnn_org','electra_org','xlnet_org','bert_org','roberta_org'], axis = 1)
final_df_pred.columns = ['Evaluation','mnb', 'cnn', 'electra', 'xlnet', 'bert', 'roberta']
final_df_pred
```

```python
final_df_org.set_index('Evaluation', inplace=True)
final_df_pred.set_index('Evaluation', inplace=True)
```

```python
# Get the test accuracy values for 'Label3' from both DataFrames
Original_df1 = final_df_org.loc['test_accuracy']
Predicted_df2 = final_df_pred.loc['test_accuracy']

 
plt.figure(figsize=(6, 4))
plt.plot(final_df_org.columns, Original_df1, marker='o', label='Original data')
plt.plot(final_df_pred.columns, Predicted_df2, marker='o', label='Translated data')
plt.xlabel('Models')
plt.ylabel('Test Accuracy')
plt.title('Test Accuracy Comparison for Original and Translated data')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Get the test accuracy values for 'Label3' from both DataFrames
Train_df1 = final_df_org.loc['train_accuracy']
Val_df1 = final_df_org.loc['val_accuracy']

plt.figure(figsize=(6, 4))
plt.plot(final_df_org.columns, Train_df1, marker='o', label='Train Accuracy')
plt.plot(final_df_pred.columns, Val_df1, marker='o', label='Val Accuracy')
plt.xlabel('Models')
plt.ylabel('Test Accuracy')
plt.title('Train and Val accuracy Comparison')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Extract precision data from both dataframes
org_precision = final_df_org.loc['precision']
pred_precision = final_df_pred.loc['precision']

org_recall = final_df_org.loc['recall']
pred_recall = final_df_pred.loc['recall']

org_f1 = final_df_org.loc['f1']
pred_f1 = final_df_pred.loc['f1']

org_kappa = final_df_org.loc['kappa']
pred_kappa = final_df_pred.loc['kappa']

# Create an array for the x-values
x = np.arange(len(org_precision)) 
width = 0.35

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(8, 8)) 
# Create the bars
ax1.bar(x - width/2, org_precision, width, label='Original')
ax1.bar(x + width/2, pred_precision, width, label='Translated')

ax2.bar(x - width/2, org_recall, width, label='Original')
ax2.bar(x + width/2, pred_recall, width, label='Translated')

ax3.bar(x - width/2, org_f1, width, label='Original')
ax3.bar(x + width/2, pred_f1, width, label='Translated')

ax4.bar(x - width/2, org_kappa, width, label='Original')
ax4.bar(x + width/2, pred_kappa, width, label='Translated')

# Set up labels, title, and legend
ax1.set_xlabel('Models')
ax1.set_ylabel('Precision')
ax1.set_title('Precision scores for original and translated data')
ax1.set_xticks(x)
ax1.set_xticklabels(final_df_org.columns)  # Assuming model names are the columns of your dataframe
ax1.legend()

ax2.set_xlabel('Models')
ax2.set_ylabel('Recall')
ax2.set_title('Recall scores for original and translated data')
ax2.set_xticks(x)
ax2.set_xticklabels(final_df_org.columns)  # Assuming model names are the columns of your dataframe
ax2.legend()

ax3.set_xlabel('Models')
ax3.set_ylabel('F1')
ax3.set_title('F1 scores for original and translated data')
ax3.set_xticks(x)
ax3.set_xticklabels(final_df_org.columns)  # Assuming model names are the columns of your dataframe
ax3.legend()

ax4.set_xlabel('Models')
ax4.set_ylabel('kappa')
ax4.set_title('Kappa values for original and translated data')
ax4.set_xticks(x)
ax4.set_xticklabels(final_df_org.columns)  # Assuming model names are the columns of your dataframe
ax4.legend()


fig.tight_layout()

plt.show()
```
